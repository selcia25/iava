```python
import matplotlib.pyplot as plt 
import cv2
```

```python
def display_image(title, image):
    plt.figure(figsize=(4, 4))
    plt.title(title)
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('church.jpg')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display_image('Original Image', rgb_image)
```

```python
gray_scale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display_image('Grayscale Image', gray_scale)
```

```python
# Edge Detection - Sobel Operator, Canny, Laplacian, Prewitt, Robert
```

```python
from skimage.filters import sobel, roberts, prewitt

sobel_edges = sobel(gray_scale)
display_image('Sobel Operator', sobel_edges)
```

```python
roberts_operator = roberts(gray_scale)
display_image('Roberts Operator', roberts_operator)
```

```python
prewitt_filter = prewitt(gray_scale)
display_image('Prewitt Filter', prewitt_filter)
```

```python
laplacian_filter = cv2.Laplacian(gray_scale, cv2.CV_16S)
display_image('Laplacian Filter', laplacian_filter)
```

```python
canny_edge_detector = cv2.Canny(gray_scale, 100, 200)
display_image('Canny Edge Detector', canny_edge_detector)
```
